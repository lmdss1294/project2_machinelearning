```python
import pandas as pd

# 데이터 파일 경로와 파일 이름을 리스트에 저장
train_files = ['https://raw.githubusercontent.com/lmdss1294/project2_machinelearning/main/data/2010.csv',
         'https://raw.githubusercontent.com/lmdss1294/project2_machinelearning/main/data/2011.csv',
         'https://raw.githubusercontent.com/lmdss1294/project2_machinelearning/main/data/2012.csv',
         'https://raw.githubusercontent.com/lmdss1294/project2_machinelearning/main/data/2013.csv'
         ]

# 데이터프레임 리스트 생성
dfs = []
for file in train_files:
    dfs.append(pd.read_csv(file))

# 데이터프레임 병합
train_data = pd.concat(dfs, ignore_index=True)

# 결과 확인
print(train_data.head())
```

```python
train_data['항공사'].fillna('개인',inplace = True)
```
