```python
df_train = train_data.copy()
df_test = test_data.copy()
```

```python
df_train['현황'] = df_train['현황'].replace({'지연': '출발', '회항': '취소'})
df_test['현황'] = df_test['현황'].replace({'지연': '출발', '회항': '취소'})
```

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
```

```python
df_train.drop(labels = '출발공항코드',axis = 1,inplace = True)
df_test.drop(labels = '출발공항코드',axis = 1,inplace = True)
```

```python
# OrdinalEncoder 객체 생성
encoder = OrdinalEncoder()

# 카테고리형 데이터 추출
cat_cols = ['항공사', '도착공항코드']
cat_train = df_train[cat_cols]
cat_test = df_test[cat_cols]

# fit_transform() 함수를 이용하여 카테고리형 데이터를 숫자형으로 변환
df_train[cat_cols] = encoder.fit_transform(cat_train)
df_test[cat_cols] = encoder.transform(cat_test)
     
```

```python
# 다수 클래스 데이터와 소수 클래스 데이터 분리
majority = df_train[df_train[target] == '출발']
minority = df_train[df_train[target] == '취소']
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 다수 클래스 데이터와 소수 클래스 데이터 분리
majority = df_train[df_train[target] == '출발']
minority = df_train[df_train[target] == '취소']

# 다수 클래스 데이터 1만개 랜덤 샘플링
majority_sampled = majority.sample(n=4000, random_state=42)

# 소수 클래스 데이터와 결합
df_resampled = pd.concat([minority, majority_sampled], axis=0)

# SMOTE 오버샘플링 적용
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(df_resampled.drop(columns=[target]), df_resampled[target])

# 학습셋과 검증셋 분리
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# 이진 분류인 경우 클래스 가중치 계산
class_weights = dict(1 / y_train.value_counts(normalize=True))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# 그래디언트 부스팅 모델 생성
gbc = GradientBoostingClassifier()

# 탐색할 하이퍼파라미터 공간 지정
param_distributions = {
    'n_estimators': randint(100, 1000),
    'learning_rate': [0.001, 0.01, 0.1],
    'max_depth': randint(1, 10),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10)
}

# 랜덤 서치 객체 생성
random_search = RandomizedSearchCV(gbc, param_distributions, n_iter=10, cv=5, scoring='f1', n_jobs=-1, random_state=42)

# 랜덤 서치 수행
random_search.fit(X_train, y_train)

```

```python
# 최적 하이퍼파라미터 확인
print('Best parameters:', random_search.best_params_)
from sklearn.metrics import f1_score
```

```python
best_gbc = random_search.best_estimator_
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

X_test = df_test.drop(labels = '현황',axis =1)
y_pred_grad = best_gbc.predict(X_test)
y_test = df_test['현황']
print(classification_report(y_test, y_pred_grad))
```

```python
from sklearn.metrics import roc_curve

# y_pred_grad는 best_gbc.predict(X_test)의 결과입니다.
y_true_binary = [1 if label == '취소' else 0 for label in y_test]
y_pred_binary = [1 if label == '취소' else 0 for label in y_pred_grad]

fpr, tpr, thresholds = roc_curve(y_true_binary, y_pred_binary, pos_label=1)

# ROC 곡선 출력
import matplotlib.pyplot as plt

plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true_binary, y_pred_binary)

print('Confusion Matrix:')
print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# confusion matrix 시각화
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
